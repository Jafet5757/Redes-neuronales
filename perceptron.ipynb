```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Perceptron:
  def __init__(self, input_size, learning_rate=0.01, epochs=10):
    self.weights = np.random.rand(input_size)
    self.bias = np.random.rand(1)
    self.learning_rate = learning_rate
    self.epochs = epochs

  def activation_function(self, x):
    return 1 if x >= 0 else 0
  
  def predict(self, x):
    return self.activation_function(np.dot(self.weights, x) + self.bias)
  
  def fit(self, X, Y):
    for epoch in range(self.epochs):
      for inputs, label in zip(X, Y):
        prediction = self.predict(inputs)
        self.weights += self.learning_rate * (label - prediction) * inputs
        self.bias += self.learning_rate * (label - prediction)
```

```python
# Datos de entrada y etiquetas para la compuerta AND
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])  # Salida deseada de la compuerta AND

# Instanciar el perceptrón
perceptron = Perceptron(input_size=2)

# Entrenar el perceptrón
perceptron.fit(X, y)

# Hacer predicciones
output = perceptron.predict([0, 0])
print(output)
```

```python
# Generar predicciones para todos los datos de entrada
predictions = np.array([perceptron.predict(x) for x in X])

# Graficar las predicciones
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=predictions, cmap='viridis', marker='o', label='Predicciones')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', marker='x', label='Datos Reales')
plt.title('Predicciones del Perceptrón vs Datos Reales')
plt.xlabel('Entrada 1')
plt.ylabel('Entrada 2')
plt.legend()
plt.show()
```
